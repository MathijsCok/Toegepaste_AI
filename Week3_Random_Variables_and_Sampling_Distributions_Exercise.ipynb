### Visualizing the Law of Large Numbers

Let’s visualize how the sample mean stabilizes as more data are collected. Adjust the code to simulate 10,000 rolls (fill in the gaps ...). Interpret the plot.

```python
# rolls = ...  /add your code here
Dice_sides = 6
amount = 100
x = np.linspace(0, amount, amount)
rolls_running = np.random.randint(1, Dice_sides + 1, size=amount)


running_means = np.cumsum(rolls_running) / np.arange(1, len(rolls_running)+1)

plt.figure(figsize=(10,5))
plt.plot(running_means, label='Running Mean')
plt.axhline(3.5, color='red', linestyle='--', label='Theoretical Mean (3.5)')
plt.title('Law of Large Numbers Demonstration')
plt.xlabel('Number of Rolls')
plt.ylabel('Running Mean')
plt.ylim(0.5, 6.5)
plt.legend()
plt.show()
```
