## Exercise 1 – Descriptive Statistics
Given the following dataset:

`data = [12, 15, 14, 10, 18, 20, 17, 16, 14, 13]`

1. Compute the mean, median, and mode.
2. Compute the variance and standard deviation.
3. Interpret the results.

```python
def find_mean(list):
    return sum(list)/len(list)

def find_median(list):
    listlength = len(list)
    isEven = False if listlength % 2 else True
    
    # sort the list
    list.sort()
    
    if isEven:
        # even has 2 in center, mean of the 2 in the center
        middleLow = list[int(listlength/2)-1]
        middleHigh = list[int(listlength/2)]
        return find_mean((middleLow, middleHigh))
    else:
        # uneven has 1 in center
        return list[int(listlength/2)]
    
def find_mode(list):
    # find all unique values
    uniques = []
    for x in list:
        if x not in uniques:
            uniques.append(x)

    highest_count = 0

    for u in uniques:
        # find amount per number
        count = list.count(u)
        
        # grab with highest amount
        if count > highest_count:
            highest_count = count
            most_frequent = [u]
        elif count == highest_count:
            most_frequent.append(u)
             
    return most_frequent
```

```python
List = [2, 1, 1, 1, 3]
mode = find_mode(List)
print(mode)
```

```python
# import data
data = [12, 15, 14, 10, 18, 20, 17, 16, 14, 13]

# mean
mean = find_mean(data)
print(f"mean is: {mean}")

# median
median = find_median(data)
print(f"median is: {median}")

# mode
mode = find_mode(data)
print(f"mode is: {mode[0]}")
```

## Exercise 2 – Discrete Random Variables

A fair six-sided die is rolled once.

1. Identify the type of distribution.
2. What is the probability that the outcome is greater than 4?


```python
def single_dice_chance(Ltresh=-1, Utresh=10^99, sides = 6):
    """
    higher then x
    lower then y
    """ 
    counter = 0
    
    for i in range(max(1, Ltresh), min(sides, Utresh)+1):
        counter += 1
    
    
    chance = counter/sides
    percentage = chance*100
    
    return chance, percentage
    
    
print(single_dice_chance(Ltresh = 5))
    
```
